```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
CCTAG_PATH = './cctag-lib/'
CCSIZE = 450
cc1 = cv2.imread(CCTAG_PATH + '0000.png')
cc2 = cv2.imread(CCTAG_PATH + '0001.png')
cc3 = cv2.imread(CCTAG_PATH + '0002.png')
cc4 = cv2.imread(CCTAG_PATH + '0003.png')

canvas = np.zeros((2480,3508,3) , dtype = int)
canvas[:] = 255

PAD = 150
lastx = 300
lasty = 300
canvas[lasty:lasty+CCSIZE, lastx:lastx+CCSIZE] = cc1
lastx = lastx + CCSIZE + PAD
canvas[lasty:lasty+CCSIZE, lastx:lastx+CCSIZE] = cc2
lasty = lasty + CCSIZE + PAD
canvas[lasty:lasty+CCSIZE, lastx:lastx+CCSIZE] = cc4
lastx = lastx - CCSIZE - PAD
canvas[lasty:lasty+CCSIZE, lastx:lastx+CCSIZE] = cc3



plt.imshow(canvas)
cv2.imwrite('3xCCTAG.png', canvas)
```
